# When data is confusing ...... - pivot table (with margin attribute

```python
df.pivot_table(index=["Country"],
          columns = ["Region"],
          values = ["Quantity","Total"],
          aggfunc=[np.sum])
```

```python
df.pivot_table(index=["Country"],
          columns = ["Region"],
          values = ["Quantity","Total"],
          aggfunc=[np.sum],
             margins=True,
              margins_name="Grand_Total")
```
