```python
x = np.linspace(0,100,5)
x
```

```python
y = np.sqrt(x)
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'r')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("X vs Y")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'b')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'g')
plt.subplot(2,2,4)
plt.plot(x,y,'r')
plt.show()
```

<h2>Object oriented method</h2>

```python
fig = plt.figure()
axes = fig.add_axes([0.2,0.2,0.7,0.7])
axes.plot(x,y)
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("X vs Y")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.55,0.4,0.3])
axes1.plot([0.1,0.5,.013,0.17],[10,15,18,25],'red')
axes2.plot(y,x,'g')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'r')
axes[1].plot(y,x,'g')

#To avoid overlapping
plt.tight_layout()
```

```python
fig= plt.figure(figsize=(3,2),dpi=100) #3,2 in inches
axd = fig.add_axes([0,0,1,1])
axd.plot(x,y)
```

<h2> Bar graphs</h2>

```python
div = ["A","B","C","D","E"]
marks = [54,4,53,23,31]
variance=[2,3,2,4,5]
plt.barh(div,marks,color='green',xerr=variance)
```

```python
div = ["A","B","C","D","E"]
marks = [54,4,53,23,31]
marks2 = [54,4,53,23,31]
index = np.arange(5)
width = 0.3

plt.bar(index,marks,width,color='green')
plt.bar(index+width,marks2,width,color='blue')
plt.xticks(index+width/2,div)
```

<h2>Pie chart </h2>

```python
plt.pie(marks,labels=div,shadow='true',startangle=45) #explode = Explode
plt.axis('equal')
```
